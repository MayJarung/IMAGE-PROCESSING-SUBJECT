```python
#Lab2.2

import cv2
import numpy as np
from matplotlib import pyplot as plt

# 
original_img = cv2.imread("img.jpg")

# convert  BGR to RGB
original_rgb = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)

#  R, G, B
r,g,b = cv2.split(original_rgb)

#  Histogram Equalization 
r_eq = cv2.equalizeHist(r)
g_eq = cv2.equalizeHist(g)
b_eq = cv2.equalizeHist(b)

#  Equalization 
equalized_img = cv2.merge((r_eq, g_eq, b_eq))

"""
     Histogram  original_img, equalized_img 
     cv2.calCHist()
"""
hist_r = cv2.calcHist([original_rgb], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([original_rgb], [1], None, [256], [0, 256])
hist_b = cv2.calcHist([original_rgb], [2], None, [256], [0, 256])

hist_r_eq = cv2.calcHist([equalized_img], [0], None, [256], [0, 256])
hist_g_eq = cv2.calcHist([equalized_img], [1], None, [256], [0, 256])
hist_b_eq = cv2.calcHist([equalized_img], [2], None, [256], [0, 256])

#  subplot
fig, axes = plt.subplots(2, 2, figsize=(15, 6))

axes[0,0].imshow(original_rgb)
axes[0,0].set_title('Original Color Image')

axes[0,1].plot(hist_r, color='r')
axes[0,1].plot(hist_g, color='g')
axes[0,1].plot(hist_b, color='b')
axes[0,1].set_title('Hist Original Image')

axes[1,0].imshow(equalized_img)
axes[1,0].set_title('Equalized Image')

axes[1,1].plot(hist_r_eq, color='r')
axes[1,1].plot(hist_g_eq, color='g')
axes[1,1].plot(hist_b_eq, color='b')
axes[1,1].set_title('Hist Equalized Image')
```
